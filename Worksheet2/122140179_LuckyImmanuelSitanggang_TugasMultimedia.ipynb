# Pendahuluan

## Apa itu Data Multimedia?

Data multimedia adalah informasi yang dikodekan dalam berbagai format untuk merepresentasikan dunia nyata:

- **Audio (1D)**: Sinyal satu dimensi yang berubah terhadap waktu
  - Contoh: musik, suara, speech
  - Representasi: amplitudo vs waktu
  
- **Gambar (2D)**: Matriks nilai intensitas dalam ruang dua dimensi
  - Contoh: foto, ilustrasi, grafik
  - Representasi: intensitas pixel pada koordinat (x,y)
  
- **Video (2D + Waktu)**: Rangkaian frame (gambar) yang ditampilkan berurutan
  - Contoh: film, rekaman, animasi
  - Representasi: frame berubah terhadap waktu dengan frame rate tertentu

### A4. TODO: Spectrogram log-dB

**Instruksi:**
Hitung STFT dan tampilkan spectrogram dalam skala log-dB:
- Gunakan parameter standar (n_fft=1024, hop_length=256)
- Tampilkan dengan colorbar
- Label sumbu: waktu (detik) dan frekuensi (Hz)

**Analisis yang diperlukan:**
Jelaskan perbedaan informasi yang didapat dari spectrogram dibanding waveform.

perbedaan paling besar adalah pada grafiknya dimana waveform sumbu y  menggambarkan amplitudo sedangkan spectogram sumbu Y nya menggambarkan energi sehingga waveform bisa melihat pada kapan musik sangat keras sedangkan spectogram melihat kapan frekuensi tinggi terjadi.


```python
y_mono = y[:,0] if y.shape[1] == 1 else y.mean(axis=1)

n_fft = 1024
hop_length = 256

# STFT ke dB
S = librosa.stft(y_mono, n_fft=n_fft, hop_length=hop_length, window = 'hann', center = True)
S_db = librosa.amplitude_to_db(np.abs(S), ref=np.max)

t_max = S_db.shape[1] * hop_length / sr
f_max = sr / 2
plt.figure(figsize=(10, 4))
im = plt.imshow(S_db, aspect='auto', origin='lower', cmap='magma',extent=[0, t_max, 0, f_max])
cbar = plt.colorbar(im, format='%+0.1f dB')
cbar.set_label('Magnitudo (dB)')
plt.title('Spectogram (log-dB)')
plt.xlabel('Waktu (detik)')
plt.ylabel('Frekuensi (Hz)')
plt.tight_layout()
plt.show()
```

```python
# Misalnya audio y sudah stereo: shape (n_samples, 2), dan sr sudah ada
L = y[:, 0]
R = y[:, 1]

n_fft = 1024
hop_length = 256

# STFT untuk kiri dan kanan
SL = librosa.stft(L, n_fft=n_fft, hop_length=hop_length, window='hann', center=True)
SR = librosa.stft(R, n_fft=n_fft, hop_length=hop_length, window='hann', center=True)

# Konversi ke dB, gunakan max dari kedua channel biar skala konsisten
ref = max(np.abs(SL).max(), np.abs(SR).max())
SL_db = librosa.amplitude_to_db(np.abs(SL), ref=ref)
SR_db = librosa.amplitude_to_db(np.abs(SR), ref=ref)

# Hitung sumbu waktu dan frekuensi
t_max = SL_db.shape[1] * hop_length / sr
f_max = sr / 2
vmin = min(SL_db.min(), SR_db.min())
vmax = 0.0  # karena ref global max

# Plot dengan colormap berbeda
fig, axs = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

im0 = axs[0].imshow(SL_db, origin='lower', aspect='auto',
                    extent=[0, t_max, 0, f_max], cmap='magma', vmin=vmin, vmax=vmax)
axs[0].set_title('Spectrogram L (dB)')
axs[0].set_xlabel('Waktu (detik)')
axs[0].set_ylabel('Frekuensi (Hz)')

im1 = axs[1].imshow(SR_db, origin='lower', aspect='auto',
                    extent=[0, t_max, 0, f_max], cmap='inferno', vmin=vmin, vmax=vmax)
axs[1].set_title('Spectrogram R (dB)')
axs[1].set_xlabel('Waktu (detik)')

# Tambahkan colorbar
cbar0 = fig.colorbar(im0, ax=axs[0], format='%+0.1f dB')
cbar0.set_label('Magnitudo (dB)')
cbar1 = fig.colorbar(im1, ax=axs[1], format='%+0.1f dB')
cbar1.set_label('Magnitudo (dB)')

plt.tight_layout()
plt.show()
```

Preprocessing ini berfungsi agar tahap preprocessing berikutnya (resize, konversi warna , dll) ridak salah langkah dan pas targetnya dan hasil visualisasi sesuai dengan apa yang diinginkan serta fitur yang ingin diubah tidak salah.

**Analisis yang diperlukan:**
Analisis: channel mana yang dominan? Bagaimana kontras gambar? Seperti apa sebaran intensitasnya?

Channel yang paling dominan adalah channel biru diikuti merah lalu hijau karena memang digrafik biru mendominasi dari 150 keatas . Lalu kontras tinggi karena sangat terlihat 2 puncak pada grafik dimana puncak pertama di bawah 50 pixel dan puncak kedua diatas 250 sehingga menciptakan kontras . Untuk intensitas sama seperti kontras dimana 50 kebawah area gelap memiliki puncak dan puncak lain 200-255 untuk bulan dan bintang.

```python
hist_R = np.ravel(cv2.calcHist([img_bgr], [2], None, [256], [0, 256]))
hist_G = np.ravel(cv2.calcHist([img_bgr], [1], None, [256], [0, 256]))
hist_B = np.ravel(cv2.calcHist([img_bgr], [0], None, [256], [0, 256]))


x = np.arange(256)
plt.figure(figsize=(8,4))
plt.plot(x, hist_R, 'r', label = 'R')
plt.plot(x, hist_G, 'g', label = 'G')
plt.plot(x, hist_B, 'b', label = 'B')

plt.xlim(0,255);plt.xlabel('Intensitas pixel (0-255)'); plt.ylabel('Frekuensi')
plt.title('Histogram Warna (RGB)');plt.legend();plt.grid(alpha=0.3, ls='--')
plt.tight_layout(); plt.show()

means= img_bgr[:,:,::-1].mean(axis=(0,1))
gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
stdY = gray.std()
level = 'rendah' if stdY < 25 else ('sedang' if stdY < 60 else 'tinggi')
pdark = (gray < 50).mean() * 100; pbright = (gray > 200).mean() * 100;pmid = 100 - pdark - pbright

print(f"Kanal dominan: {'RGB'[np.argmax(means)]} (R = {means[0]:.1f}, G = {means[1]:.1f}, B = {means[2]:.1f})")
print(f"Kontras (std grayscale): {stdY:2f} â†’ {level} ")
items = [("<50 (gelap)", pdark), ("50â€“200", pmid), (">200 (terang)", pbright)]
for k, v in sorted(items, key=lambda x: x[1], reverse=True):
    print(f"- {k}: {v:.1f}%")
```

- Klasifikasi resolusi: mempermudah pemetaan ke standar distribusi (SD/HD/Full HD/4K).

Catatan: fps yang terdeteksi dari container bisa tidak akurat pada video VFR; validasi dengan time stamps jika diperlukan.


```python
video_path = 'data/guitar play_all falls down.mp4'

cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise RuntimeError(f'Gagal membuka video: {video_path}')

width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = cap.get(cv2.CAP_PROP_FPS) or 0.0
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

duration = frame_count / fps if fps and fps > 0 else None

def classify_resolution(w, h):
    v = min(w, h)  
    if v >= 4320:
        return '8K UHD (4320p)'
    elif v >= 2160:
        return '4K UHD (2160p)'
    elif v >= 1440:
        return '2K/QHD (1440p)'
    elif v >= 1080:
        return 'Full HD (1080p)'
    elif v >= 720:
        return 'HD (720p)'
    elif v >= 480:
        return 'SD (480p)'
    else:
        return 'Sub-SD (<480p)'

res_class = classify_resolution(width, height)

print('File:', video_path)
print(f'Resolusi: {width} x {height} px')
print(f'Frame rate: {fps:.3f} fps' if fps and fps>0 else 'Frame rate: Tidak diketahui')
print(f'Jumlah frame: {frame_count}')
print(f'Durasi: {duration:.2f} detik' if duration is not None else 'Durasi: Tidak dapat dihitung (fps=0)')
print('Klasifikasi resolusi: ',classify_resolution(width, height))

# cap.release()

```

### C3. TODO: Tampilkan 3 Frame (Awalâ€“Tengahâ€“Akhir)

**Instruksi:**
Ambil dan tampilkan 3 frame representatif:
- Frame pertama (index 0)
- Frame tengah (index ~total_frame/2)
- Frame terakhir (index total_frame-1)
- **Konversi BGRâ†’RGB** sebelum ditampilkan
- Subplot dengan judul frame dan timestamp

**Analisis yang diperlukan:**
Deskripsikan perbedaan visual antar frame dan apa yang dapat dipelajari dari sampel frame ini.

Perbedaan paling jelas awal video dipenuhi dengan blackscreen , dan hanya frame tengah yang menunjukan konten dari video ini serta akhir video menunjukan layar hiram juga namun ada siluet konten dari video yang merupakan transisi pengakhiran video.


```python
n = frame_count
idxs = [0, n//2 if n > 0 else 0, max(0, n-1)]

def grab_rgb(i):
    cap.set(cv2.CAP_PROP_POS_FRAMES, i)
    ok, bgr = cap.read()
    if not ok or bgr is None:
        return None
    return cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

frames = [grab_rgb(i) for i in idxs]

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, fr, i in zip(axes, frames, idxs):
    ax.axis('off')
    if fr is None:
        ax.set_title(f'Frame {i} â€” gagal')
    else:
        ax.imshow(fr)
        ts = (i / fps) if ('fps' in globals() and fps and fps > 0) else None
        ax.set_title(f'Frame {i} â€” {ts:.2f}s' if ts is not None else f'Frame {i}')

title = os.path.basename(video_path) if 'video_path' in globals() else 'Video'
plt.suptitle(title)
plt.tight_layout()
plt.show()
```
